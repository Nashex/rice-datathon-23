```python
import pandas as pd
import numpy as np
import cv2
import os
import matplotlib.pyplot as plt
from matplotlib import image as mpimg
from sklearn.model_selection import train_test_split
import re


# Load in the users data
users = pd.read_csv("./Users.csv")

def getOcrFiles():
    ocr = os.listdir("./new_ocr")
    return ocr

def readOcrFile(fileId):
    try:
        df = pd.read_csv(f"./new_ocr/{fileId}.csv")
    
        if not len(df):
            return None

        df["1"] = df["0"].apply(lambda x: re.sub("[^A-Za-z0-9:\\/\s\.,]", "", re.sub("[\n\s]+", " ", str(x).upper())).strip())
        df["0"] = fileId
        return df
    except:
        return None

def showImg(fileId):
    plt.title(fileId)
    image = mpimg.imread("./img/" + fileId + ".jpg")
    plt.imshow(image)
    plt.show()
```

```python
ocrFiles = getOcrFiles()

def filter_price(row):
    # Match a price using regex
    match = re.match(r'[\-\$)]?\d+[.]\d*', str(row[row.index[-1]]))
    if match:
        return match.group().replace("$", "")
    else:
        return None
    
def processFilePrices(ocrFile):
    fileId = ocrFile.split(".")[0]
    df = readOcrFile(fileId)

    # Apply the filter_price function to the dataframe
    df['price'] = df.apply(filter_price, axis=1)
    df.dropna()
    return list(df['price'].dropna())

res = pd.DataFrame()
for fid, prices in [(f.split(".")[0], processFilePrices(f)) for f in ocrFiles]:
    res = res.append([[fid, prices]])

res.to_csv("price_test.csv", index=False)
#print(ocrFile)
#readOcrFile(ocrFile)
```

```python
ocrFiles = getOcrFiles()

def processFile(ocrFile):
    fileId = ocrFile.split(".")[0]

    if (not fileId):
        return None
    df = readOcrFile(fileId)
    
    
    return " ".join(df[df.columns[-1]].apply(lambda x: str(x).upper()))
  
res = pd.DataFrame()
for file in ocrFiles:
    fid = file.split(".")[0]
    df = readOcrFile(fid)
    if df is not None:
        res = res.append([df])
    
train, test = train_test_split(res, test_size=0.2, random_state=200)
#train.to_csv("ocr_train.csv")
#test.to_csv("ocr_test.csv")
```
